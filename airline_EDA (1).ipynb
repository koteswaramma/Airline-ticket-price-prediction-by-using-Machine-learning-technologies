```python
col_cat=['Airline','Source','Destination','Total_Stops','Additional_Info']
```

```python
@interact
def plot(col=col_cat):
    train[str(col)].value_counts().plot.barh()
    plt.show()
```

```python
@interact
def plot(col1=col_cat,col2=col_cat):
    train.groupby([col1,col2])['Price'].count().sort_values(ascending=False).plot.barh()
    plt.show()
```

```python
train.Route.value_counts()[0:10].plot.barh()
plt.show()
train.Route.value_counts()[10:20].plot.barh()
plt.show()
```

```python
@interact
def plot(col1=col_cat):
    sns.boxplot(data=train,x=str(col1),y='Price')
    plt.xticks(rotation=45)
    plt.show()
```
